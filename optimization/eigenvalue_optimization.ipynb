#  What is the relation between quadratic form, Hessian, Mahalanobis distance and optimization?


Hessian:The Hessian is the second derivatives and second-order cross partials of a scalar function taken at a point.

$\mathbf{H}=\left[\begin{array}{cccc}{\frac{\partial^{2} f}{\partial x_{1}^{2}}} & {\frac{\partial^{2} f}{\partial x_{1} \partial x_{2}}} & {\cdots} & {\frac{\partial^{2} f}{\partial x_{1} \partial x_{n}}} \\ {\frac{\partial^{2} f}{\partial x_{2} \partial x_{1}}} & {\frac{\partial^{2} f}{\partial x_{2}^{2}}} & {\cdots} & {\frac{\partial^{2} f}{\partial x_{2} \partial x_{n}}} \\ {\vdots} & {\vdots} & {\ddots} & {\vdots} \\ {\frac{\partial^{2} f}{\partial x_{n} \partial x_{1}}} & {\frac{\partial^{2} f}{\partial x_{n} \partial x_{2}}} & {\cdots} & {\frac{\partial^{2} f}{\partial x_{n}^{2}}}\end{array}\right]$
<p>The idea is that the function ∆í has many inputs x‚ÇÅ, x‚ÇÇ, x‚ÇÉ, ‚Ä¶ which all affect its one scalar output. By changing the xi's you change that output. You can't write just ∆í‚Ä≤ because that wouldn't specify which of the xi is being changed. (You need the Jacobian.) And you can't write just ∆í‚Ä≥ for the same reason. So you need the Hessian.</p>
 <p>   
the second-order derivatives represent the curvature of the function. The difference is that the curvature will differ according to the direction, which presents a whole new set of problems.
</p>
 <p>A two-dimensional quadratic form is simply any function that can be expressed like:</p>
$\frac{1}{2}a_{11}x_1^2 + a_{12}x_1x_2 +\frac{1}{2}a_{22}x_2^2 + b_1x_1 + b_2x_2 + c  $
<p>When we write this in matrix form, we get the following equation
$ \frac{1}{2}x^TAx - b^Tx + c $ 
where A is symmetric (its i -th row and i -th column are the same)
the gradient is Ax - b  and the Hessian is A .

 The eigenvalues and eigenvectors of the Hessian are going to be the bridge between various equations and analysis surrounding the Hessian and our intuition.
 In the case of the Hessian, the eigenvectors and eigenvalues have the following important properties:

<p>-Each eigenvector represents a direction where the curvature is independent of the other directions</p>
<p>-The curvature in the direction of the eigenvector is determined by the eigenvalue. If the eigenvalue is larger, there is a larger curvature, and if it positive, the curvature will be positive, and vice-versa.</p>

<P><b>The relation between the Hessian and optimization</p></b>
<p>the larger the eigenvalue, the faster the convergence from the direction of its corresponding eigenvector.What concerns us is whether the sign of the gradient is the same. <p>the step-size is effectively bounded by the largest curvature = largest eigenvalue.</p>
<p>the speed of convergence is determined by the smallest eigenvalue.</p>
<p> the ratio between the largest and smallest eigenvalues in the Hessian is a very important value and has its own name: the ‚Äúcondition number‚Äù. The problem of the condition number being very large is called ‚Äúill-conditioning‚Äù, and is a prevalent problem in many areas of optimization.</p>


#  What is momentum and what problem is it solving? You may use this as an additional reference.


### Momentum methods in the context of the machine learning refer to a group of trick and techniques designed to speed up convergence of first order optimization method like gradient descent.
They essentially work by adding what‚Äôs called the momentum term to the update formula for gradient descent, thereby ameliorating its natural ‚Äúzigzagging behavior,‚Äù especially in long narrow valleys of the cost function.

$W‚Ä≤_i=W_i‚àíŒ±ŒîWi+ŒºŒîW_i‚àí_1$ 
where Œ± is the learning rate, and Œº is the momentum term.

if the Œº term is larger than the Œ± term then in the next iteration the ŒîW from the previous iteration will have a greater influence on the weight than the current one.
Can be used to handle noisy gradients
Can handle extremely small gradients

Using Backpropagation with momentum in a network with n different weights Wk the i-th correction for weight Wk is given by

$ŒîW_k(i)=‚àíŒ±‚àÇE/‚àÇWk+ŒºŒîW_k(i‚àí1)$ where $‚àÇE/‚àÇWk$ is the variation of the loss w.r.t. $Wk$.

Introduction of the momentum rate allows the attenuation of oscillations in the gradient descent. The geometric idea behind this idea can probably best be understood in terms of an eigenspace analysis in the linear case. If the ratio between lowest and largest eigenvalue is large then performing a gradient descent is slow even if the learning rate large due to the conditioning of the matrix. The momentum introduces some balancing in the update between the eigenvectors associated to lower and larger eigenvalues.

```python
f  = lambda x, y: (1.5 - x + x*y)**2 + (2.25 - x + x*y**2)**2 + (2.625 - x + x*y**3)**2
```

```python
xmin, xmax, xstep = -4.5, 4.5, .2
ymin, ymax, ystep = -4.5, 4.5, .2
x, y = np.meshgrid(np.arange(xmin, xmax + xstep, xstep), np.arange(ymin, ymax + ystep, ystep))
z = f(x, y)
minima = np.array([3., .5])
minima_ = minima.reshape(-1, 1)
f(*minima_)
```

```python
x0 = np.array([3., 4.])
func = value_and_grad(lambda args: f(*args))
res = minimize(func, x0=x0, method='Newton-CG',
               jac=True, tol=1e-20, callback=print)
```
